```python
from utils import authenticate_and_initialize
authenticate_and_initialize()
```

```python
from utils import load_panama_geometry
pa_geometry = load_panama_geometry()
```

```python
from utils import get_embeddings_collection, get_year_embedding
embeddings = get_embeddings_collection(pa_geometry)
```

```python
emb24 = get_year_embedding(2024, embeddings)
emb23 = get_year_embedding(2023, embeddings)
```

```python
from utils import sample_embeddings, train_kmeans_clusterer, create_cluster_image

# Sample embeddings for training
sample = sample_embeddings(
    emb24,
    region=pa_geometry,
    scale=50,
    num_pixels=5000,
    seed=42,
    geometries=False
)

# Train K-Means clusterer
clusterer = train_kmeans_clusterer(sample, n_clusters=8)

# Create clustered image
clusters = create_cluster_image(emb24, clusterer, clip_geometry=pa_geometry)
```

```python
from utils import create_geemap, VIS_CLUSTER_8

# Create and configure map
Map = create_geemap(center_geometry=pa_geometry, zoom=8)
Map.addLayer(clusters, VIS_CLUSTER_8, "Clústeres latentes")
Map
```

```python
from utils import export_thumbnail

url = export_thumbnail(
    clusters,
    pa_geometry,
    fname="panama_2024_clusters",
    vis_params=VIS_CLUSTER_8,
    dimensions=2048,
    print_url=True
)
```
